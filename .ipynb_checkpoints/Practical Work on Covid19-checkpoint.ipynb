## Objective 3
### Using pandas, display the “taux_occupation_sae” > 50

```python
taux_occupation_saes_more_than_50 = df[df['taux_occupation_sae'] > 50]
taux_occupation_saes_more_than_50.head()
```

## Objective 4
### Calculate the mean, min and max of taux_occupation_sae

```python
taux_occupation_sae_describe = df['taux_occupation_sae'].describe()
taux_occupation_sae_mean = taux_occupation_sae_describe['mean']
taux_occupation_sae_min = taux_occupation_sae_describe['min']
taux_occupation_sae_max = taux_occupation_sae_describe['max']
print(f'Details of taux_occupation_sae Column: mean: {taux_occupation_sae_mean}, min: {taux_occupation_sae_min}, max: {taux_occupation_sae_max}')
```

## Objective 5
### Calculate the mean, min and max of tx_incid

```python
tx_incid_mean = df['tx_incid'].mean()
tx_incid_min = df['tx_incid'].min()
tx_incid_max = df['tx_incid'].max()
print(f'Details of tx_incid Column: mean: {tx_incid_mean}, min: {tx_incid_min}, max: {tx_incid_max}')
```

## Objective 6
### Write the results of (4) and (5) in a csv file named TP2-Covid19-stat.csv

```python
datas = {
    'taux_occupation_sae_mean': [taux_occupation_sae_mean],
    'taux_occupation_sae_min' : [taux_occupation_sae_min],
    'taux_occupation_sae_max' : [taux_occupation_sae_max],
    'tx_incid_mean': [tx_incid_mean],
    'tx_incid_min': [tx_incid_min],
    'tx_incid_max': [tx_incid_mean]
}
TP2_Covid19_stat = pd.DataFrame(datas)
TP2_Covid19_stat.to_csv('TP2-Covid19-stat.csv')
```

### Using DictReader: display all the values for “taux_occupation_sae_couleur”=”rouge” (red regions)

```python
def only_answers(row):
    return 'rouge' in row.values()

dict_reader_filter = []
for row in filter(only_answers, df_dict_reader):
    dict_reader_filter.append(row)
```
